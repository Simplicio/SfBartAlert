```python
index = Temp.index[::-1].values

sum_index_set = set()

#Loop over all alerts during the day
for i in range(len(index)-1):
     
    #Check if i has been explored
    if (index[i] in sum_index_set):
        continue

    matching_index = []
    matching_index.append(index[i])
    sum_index_set.add(index[i])
    

    while (i < len(index)-1):
        
        Location = Temp.LocationSet[index[i]]
        Cause = Temp.CauseSet[index[i]]
    
        FoundMatchJ = -1
        for j in range(i+1,len(index)):
            
            #Check if j has been explored
            if (index[j] in sum_index_set):
                continue
            
            #Check that two tweets are within an 1.5 hour
            deltaT = (Temp.TimeStamp[index[j]] - Temp.TimeStamp[index[i]]).total_seconds()
            #print(deltaT)
            if (deltaT > 90*60):
                break
        
            SecondLocation = Temp.LocationSet[index[j]]
            SecondCause = Temp.CauseSet[index[j]]

            U_loc = Location & SecondLocation
            I_loc = Location | SecondLocation
            
            U_cause = Cause & SecondCause
                     
            #Either find a match in the cause, or the second tweet has no cause
            if (len(U_cause) == 0):
                if not((Temp.NoCause[index[i]] == True) or (Temp.NoCause[index[j]] == True)):
                    
                    #As an exception, if all stations mentioned has a complete match, and within 15 min
                    
                    if not((Temp.TimeStamp[index[j]] - Temp.TimeStamp[index[i]]).total_seconds() <= 15*60):
                        continue
                    
                    #if not(len(I_loc) >=3 and (len(U_loc)==len(I_loc))):
                    if not((len(U_loc) == len(Location)) or (len(U_loc) == len(SecondLocation))):
                        continue
                        
            #Possible match, time to match location
 
            #No system wide failure
            if not((Temp.SystemWide[index[i]] == True) or (Temp.SystemWide[index[j]] == True)):

                #Require at least 75% of match in either set
                if not((len(U_loc) >= 0.75*len(Location)) or (len(U_loc) >= 0.75*len(SecondLocation))):              
                    continue
                #For few locations, require a more stringent check
                #if (len(I_loc) < 4):
                    #Check if it is within 15 min
                #    if not((Temp.TimeStamp[index[j]] - Temp.TimeStamp[index[i]]).total_seconds() <= 15*60):
                #        if not(len(U_loc) <= len(I_loc)-1):
                #            continue
                    #Not within 15 min
                #    else:
                #        if (len(U_loc) < len(I_loc)):
                #            continue
                
                #Require a 3/4 match if there are 4 or more stations
                #else:
                #    if not((len(U_loc) >= 0.75*len(Location)) or (len(U_loc) >= 0.75*len(SecondLocation))):              
                #        continue
            
            #Found a candidate match!
            FoundMatchJ = j
            break
            
        #Finish searching through j loop
        
        if FoundMatchJ > 0:
            matching_index.append(index[j])
            sum_index_set.add(index[j])
            i=j
        else:
          
            break
   
    print(matching_index)
    
if (index[len(index)-1]) not in sum_index_set:
    matching_index = [index[len(index)-1]]
    print(matching_index)
```
